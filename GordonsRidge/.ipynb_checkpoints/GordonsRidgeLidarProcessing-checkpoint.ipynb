```python
# defining a function to fill NaN values with a linear interpolation

def mask_knans(a, x):
    a = np.asarray(a)
    k = a.size
    n = np.append(np.isnan(a), [False] * (x - 1))
    m = np.empty(k, np.bool8)
    m.fill(True)

    s = n.strides[0]
    i = np.where(strided(n, (k + 1 - x, x), (s, s)).all(1))[0][:, None]
    i = i + np.arange(x)
    i = pd.unique(i[i < k])

    m[i] = False

    return m
```

```python
# Creating a copy of the dataframe and applying linear interpolation function at each height

GorWSdfIntp=GorWSdf

# 15 is used as the x value in when using the function because interpolation occurs if there is a usable point within 15x 2-minute intervals (within 30 min)
GorWSdfIntp['40m Wind Speed (m/s)']=GorWSdf['40m Wind Speed (m/s)'].interpolate().where(mask_knans(GorWSdf['40m Wind Speed (m/s)'].tolist(),15))
GorWSdfIntp['60m Wind Speed (m/s)']=GorWSdf['60m Wind Speed (m/s)'].interpolate().where(mask_knans(GorWSdf['60m Wind Speed (m/s)'].tolist(),15))
GorWSdfIntp['80m Wind Speed (m/s)']=GorWSdf['80m Wind Speed (m/s)'].interpolate().where(mask_knans(GorWSdf['80m Wind Speed (m/s)'].tolist(),15))
GorWSdfIntp['100m Wind Speed (m/s)']=GorWSdf['100m Wind Speed (m/s)'].interpolate().where(mask_knans(GorWSdf['100m Wind Speed (m/s)'].tolist(),15))
GorWSdfIntp['120m Wind Speed (m/s)']=GorWSdf['120m Wind Speed (m/s)'].interpolate().where(mask_knans(GorWSdf['120m Wind Speed (m/s)'].tolist(),15))
GorWSdfIntp['140m Wind Speed (m/s)']=GorWSdf['140m Wind Speed (m/s)'].interpolate().where(mask_knans(GorWSdf['140m Wind Speed (m/s)'].tolist(),15))
GorWSdfIntp['160m Wind Speed (m/s)']=GorWSdf['160m Wind Speed (m/s)'].interpolate().where(mask_knans(GorWSdf['160m Wind Speed (m/s)'].tolist(),15))
GorWSdfIntp['180m Wind Speed (m/s)']=GorWSdf['180m Wind Speed (m/s)'].interpolate().where(mask_knans(GorWSdf['180m Wind Speed (m/s)'].tolist(),15))
GorWSdfIntp['200m Wind Speed (m/s)']=GorWSdf['200m Wind Speed (m/s)'].interpolate().where(mask_knans(GorWSdf['200m Wind Speed (m/s)'].tolist(),15))

```

```python
# Creating copy of interpolated dataframe and resampling the 2-minute outputs to an hourly mean for comparison to hourly model data

GorWSdfIntpHr=GorWSdfIntp
GorWSdfIntpHr['Timestamp (end of interval)']=pd.to_datetime(GorWSdfIntpHr['Timestamp (end of interval)']) # converting index to datetime
GorWSdfIntpHr.set_index('Timestamp (end of interval)',inplace=True)

# taking hourly mean at each height 
GorWS40IntpHr=GorWSdfIntpHr['40m Wind Speed (m/s)'].resample('1H').mean()
GorWS60IntpHr=GorWSdfIntpHr['60m Wind Speed (m/s)'].resample('1H').mean()
GorWS80IntpHr=GorWSdfIntpHr['80m Wind Speed (m/s)'].resample('1H').mean()
GorWS100IntpHr=GorWSdfIntpHr['100m Wind Speed (m/s)'].resample('1H').mean()
GorWS120IntpHr=GorWSdfIntpHr['120m Wind Speed (m/s)'].resample('1H').mean()
GorWS140IntpHr=GorWSdfIntpHr['140m Wind Speed (m/s)'].resample('1H').mean()
GorWS160IntpHr=GorWSdfIntpHr['160m Wind Speed (m/s)'].resample('1H').mean()
GorWS180IntpHr=GorWSdfIntpHr['180m Wind Speed (m/s)'].resample('1H').mean()
GorWS200IntpHr=GorWSdfIntpHr['200m Wind Speed (m/s)'].resample('1H').mean()

# concatenating each height series into a single dataframe that displays hourly data
GorWSdfIntpHr=pd.concat((GorWS40IntpHr,GorWS60IntpHr,GorWS80IntpHr,GorWS100IntpHr,GorWS120IntpHr,GorWS140IntpHr,GorWS160IntpHr,GorWS180IntpHr,GorWS200IntpHr),axis=1)
```

```python
GorWSdf.to_csv('GordonsRidgeLidar2min')
GorWSdfIntp.to_csv('GordonsRidgeLidarInterpolated2min')
GorWSdfIntpHr.to_csv('GordonsRidgeLidarInterpolatedHourly')
```
